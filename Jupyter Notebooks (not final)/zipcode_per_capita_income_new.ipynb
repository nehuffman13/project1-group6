```python
import numpy as np
import pandas as pd
import csv
from census import Census
from config import (api_key)
c = Census(api_key, year=2018)
```

```python
chicago_zip = []

with open('Chicago Home Values Clean by Zip.csv', mode = 'r') as f:
    reader = csv.reader(f, delimiter=",")
    for n, row in enumerate(reader):
        if not n:
            continue
        chicago_zip.append(row[0])
        
#chicago_zip=list(map(int,chicago_zip)) #convert string to int
```

```python
house_csv=pd.read_csv('Chicago Home Values Clean by Zip.csv')
main_data_pd=pd.DataFrame(house_csv)
```

```python
census_data = c.acs5.get(("B19301_001E"), {'for': f'zip code tabulation area:{",".join(chicago_zip)}'})
```

```python
income_pd=pd.DataFrame(census_data)
income_pd=income_pd.rename(columns={"B19301_001E":"Per Capita Income 2018",
                                    "zip code tabulation area": "RegionName"})
```

```python
income_pd['RegionName']=income_pd['RegionName'].astype(str).astype(int) #converted an object to a int
```

```python
merge_data01= pd.merge(main_data_pd, income_pd, how='right', on='RegionName')
```

```python
merge_data_pd=pd.DataFrame(merge_data01)
merge_data_pd
```

```python
merge_data_pd=merge_data_pd.rename(columns={"4/30/2019":"Property values 4/2019",
                                   "4/30/2020":"Property values 4/2020",
                                    "RegionName":"zipcode"})
```

```python
crime_csv=pd.read_csv('zip_crime_data.csv')
crime_pd=pd.DataFrame(crime_csv)
```

```python
crime_property=pd.merge(crime_pd, merge_data_pd, how="inner", on= "zipcode")

crime_property=crime_property.drop('Unnamed: 0', 1)
crime_property=crime_property.drop('RegionType', 1)
crime_property=crime_property.drop('Metro', 1)
```

```python
crime_property.to_csv("crime_property.csv")
```

```python
crime_time_of_day=crime_property["HrOfDate"].value_counts()
crime_zip_count=crime_property["zipcode"].value_counts()
crime_time_of_day
```
