```python
# crime_data_to_load = "crime_sample.csv"
# crime_data_to_load = "Crimes2018April.csv"
crime_data_to_load = "zip_crime_data.csv"
crime_data = pd.read_csv(crime_data_to_load)
clean_crime_data = crime_data.dropna()
clean_crime_data.head(12)
```

```python
byZip_Analysis_df = \
    clean_crime_data.groupby("zipcode").agg(
        Count=('ID', 'count'))
byZip_Analysis_df.reset_index(inplace=True)
byZip_Analysis_df = byZip_Analysis_df.rename(columns = {'index':'Zip_Code'})
byZip_Analysis_df
```

```python
byZip_Analysis_df = pd.merge(byZip_Analysis_df, clean_crime_data, on="zipcode")
dropna_byZip_Analysis_df = byZip_Analysis_df.dropna()
dropna_byZip_Analysis_df = dropna_byZip_Analysis_df.sort_values('zipcode', ascending=False)
dropna_byZip_Analysis_df.drop_duplicates(subset='zipcode', keep='first', inplace=True)
dropna_byZip_Analysis_df
```
